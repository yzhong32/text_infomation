## 3. Document Relevance with Vector Space TF-IDF Model and Hadoop’s Reducer

```python
def read_and_reconstruct(file_path):
    word_doc = {}

    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            word, data = line.split(': ', 1)
            count, docids = data.strip('[]').split(', ', 1)
            count = int(count)
            docids = set(map(int, docids.strip('{}').split(', ')))
            word_doc[word] = [count, docids]

    return word_doc

file_path = '/root/testout/part-00000'
word_doc = read_and_reconstruct(file_path)

wc = [(k, v[0]) for k, v in word_doc.items()]
wc.sort(key=lambda x: x[1], reverse=True)
wc = wc[:200]

vocab = [w[0] for w in wc]
inverted_index = {w: word_doc[w] for w in vocab}
V = set(vocab)
print(wc)
print('vocabulary:', vocab)
print('inverted index:', inverted_index)
```

```python
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
import string

def preprocess_text(text):
    text = text.strip()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.lower()
    for x in text:
        if x in punctuations:
            text = text.replace(x, " ")
    words = text.split()
    stop_words = set(stopwords.words('english'))
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words if stemmer.stem(word) in V]
    return ' '.join(words)

data_path = '/content/train.csv'
test_data_path = '/content/test.csv'

punctuations = "!()-[]{};:'\',<>./?@#$%^&*_~"

df = pd.read_csv(data_path, names=['class', 'title', 'description'])
df['description'] = df['description'].apply(preprocess_text)

test_df = pd.read_csv(test_data_path, names=['class', 'title', 'description'])
test_df['description'] = test_df['description'].apply(preprocess_text)

df_origin = pd.read_csv(data_path, names=['class', 'title', 'description'])
test_df_origin = pd.read_csv(test_data_path, names=['class', 'title', 'description'])
```

```python
queries_origin = ["olympic gold athens", "reuters stocks friday", "investment market prices"]
queries = []
stemmer = PorterStemmer()

for query in queries_origin:
    wds = query.split()
    queries.append(' '.join([stemmer.stem(wd) for wd in wds]))

print(queries)
```

# 3.1

```python
def compute_relevance_TF_IDF(query, doc, N, avgdl, k=1.2, b=0.75):
    q_wds = query.split()
    d_wds = doc.split()

    score = 0
    for w in set(q_wds) & set(d_wds):
        freq = d_wds.count(w)
        tf = (freq * (k + 1)) / (freq + k * (1 - b + b * len(d_wds) / avgdl))
        f = len(inverted_index[w][1]) / N
        idf = np.log((N + 1) / f)
        score += tf * idf

    doc_length = len(d_wds)
    if doc_length > 0:
        score /= doc_length
    return score
```

```python
def process_TF_IDF(queries, queries_origin, docs, docs_origin, count, k=1.2, b=0.75):
    N = len(docs)
    avgdl = sum([len(d.split()) for d in docs]) / N
    stemmer = PorterStemmer()

    for idx, query in enumerate(queries):
        wds = query.split()
        scores = []
        for i, doc in enumerate(docs):
            scores.append((i, compute_relevance_TF_IDF(query, doc, N, avgdl, k=k, b=b)))
        scores.sort(reverse=True, key=lambda x: x[1])
        print(f"Query: {queries_origin[idx]}")
        print(f"Top {count}:")
        for i, score in scores[:count]:
            print(f"Doc id: {i+1}, Score: {score}, Description: {docs_origin.iloc[i]}")
        print(f"Bottom {count}:")
        for i, score in scores[-count:]:
            print(f"Doc id: {i+1}, Score: {score}, Description: {docs_origin.iloc[i]}")
        print('\n')
```

##4. Probabilistic Text Retrieval

#4.1

```python
def compute_relevance_prob(query, doc, C, lamb = 0.3):
    d_wds = doc.split()
    q_wds = query.split()

    score = 0
    for w in set(d_wds) & set(q_wds):
        c_wq, c_wd = q_wds.count(w), d_wds.count(w)
        d = len(d_wds)
        pwc = inverted_index[w][0] / C
        score += c_wq * np.log(1 + (1 - lamb)/lamb * c_wd / d / pwc)

    return score
```

```python
def process_prob(queries, queries_origin, docs, docs_origin, count, C):
    stemmer = PorterStemmer()

    for idx, query in enumerate(queries):
        wds = query.split()
        scores = []
        for i, doc in enumerate(docs):
            scores.append((i, compute_relevance_prob(query, doc, C)))
        scores.sort(reverse=True, key=lambda x: x[1])
        print(f"Query: {queries_origin[idx]}")
        print(f"Top {count}:")
        for i, score in scores[:count]:
            print(f"Doc id: {i+1}, Score: {score}, Description: {docs_origin.iloc[i]}")
        print(f"Bottom {count}:")
        for i, score in scores[-count:]:
            print(f"Doc id: {i+1}, Score: {score}, Description: {docs_origin.iloc[i]}")
        print('\n')
```

## 5 Simple Web Crawler

```python
import re

def clean_text(text):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    text = text.strip()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.lower()
    for x in text:
        if x in punctuations:
            text = text.replace(x, " ")
    return ' '.join(text.split())
```

```python
import requests
from bs4 import BeautifulSoup
import os

base_url = 'https://en.wikipedia.org'
main_page = '/wiki/Main_Page'

output_filename = 'downloaded_content.txt'

try:
    response = requests.get(base_url + main_page)
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.select('a[href^="/wiki/"]:not([href*=":"])')

    successful_downloads = 0

    with open(output_filename, 'w', encoding='utf-8') as outfile:
        for link in links:
            if successful_downloads >= 10:
                break

            href = link.get('href')
            if href:
                full_url = base_url + href
                try:
                    page_response = requests.get(full_url)
                    page_soup = BeautifulSoup(page_response.content, 'html.parser')

                    text = page_soup.find('div', id='bodyContent').get_text(separator='\n', strip=True)

                    outfile.write(text)
                    outfile.write("\n")

                    successful_downloads += 1

                except Exception as e:
                    print(f"Failed to download or write content for {full_url}: {e}")
except Exception as e:
    print(f"Failed to crawl base_url: {e}")
```

## 6 Document Relevance with Vector Space

```python
file_path = '/root/testout/part-00000'
word_doc = read_and_reconstruct(file_path)

wc = [(k, v[0]) for k, v in word_doc.items()]
wc.sort(key=lambda x: x[1], reverse=True)
wc = wc[:200]

vocab = [w[0] for w in wc]
inverted_index = {w: word_doc[w] for w in vocab}
V = set(vocab)
```

```python
import numpy as np
from sklearn.model_selection import KFold

def optimize_k(queries, docs, k_values):
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    optimal_ks = []

    for train_index, test_index in kf.split(docs):
        best_score = float('-inf')
        best_k = None
        N = len(docs)
        avgdl = np.mean([len(doc.split()) for doc in docs])

        for k in k_values:
            fold_scores = []

            for i in test_index:
                doc_scores = []
                for query in queries:
                    score = compute_relevance_TF_IDF(query, docs.iloc[i], N, avgdl, k=k)
                    doc_scores.append(score)
                fold_scores.append(np.mean(doc_scores))

            avg_score = np.mean(fold_scores)

            if avg_score > best_score:
                best_score = avg_score
                best_k = k

        optimal_ks.append(best_k)
        print(f"Optimal k for this fold: {best_k} with average score: {best_score}")

    return optimal_ks

k_values = np.linspace(1.2, 2.0, num=9)
optimal_ks = optimize_k(queries, df['description'], k_values)
print(optimal_ks)
```
